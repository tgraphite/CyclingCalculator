```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("VAM.csv")
grouped = df.groupby('grad')
fig, ax = plt.subplots(figsize=(4, 3))

for grad, sub_df in grouped:
    if grad == 0:
        continue
    ax.plot(sub_df.power, sub_df.VAM, label=f"{grad:.0f}")

ax.legend()

plt.grid()
plt.show()
```
