```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from scipy.signal import savgol_filter
```

```python
file = "tests/145k.csv"
df = pd.read_csv(file)

pwm = df.power_w.to_numpy(dtype=float)
pred = df.estimated_power_w.to_numpy(dtype=float)
alt = df.altitude_m.to_numpy(dtype=float)
pwm_avg = np.mean(pwm)
pred_avg = np.mean(pred)

fig, ax = plt.subplots(1, figsize=(6, 4))

y_exp = pwm[:8000]
y_pred = pred[:8000] #savgol_filter(pred, 40, 3)
avg_exp = np.mean(y_exp)
avg_pred = np.mean(y_pred)

ax.plot(y_exp, label=f"powermeter, avg={avg_exp:.1f}W ", alpha=0.8)
ax.plot(y_pred, label=f"predicted, avg={avg_pred:.1f}W", alpha=0.8)
ax.set_title('Predicted Power vs Actual Power')
ax.legend()
```
